```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from criticalZoom import order_checker
from timebudget import timebudget
timebudget.set_quiet(False)  # don't show measurements as they happen
timebudget.report_at_exit(True)  # Generate report when the program exits
```

```python
tmin=118.15
tmax=118.19
numtemp=100



#light quark mass
ml=38

#chemical potential
mu=1500

lambda1= -4.25 #parameter for mixing between dilaton and chiral field

minsigma=300
maxsigma=400

mu0=430
mu1=830
mu2=176

tmin,tmax,minsigma,maxsigma,order,temps,truesigma,Tc=order_checker(tmin,tmax,numtemp,minsigma,maxsigma, ml, mu, lambda1,mu0,mu1,mu2)
```

```python
#plot all the sigma values for each iteration
#get the standard colors for matplotlib
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

plt.scatter(temps,(truesigma[:,0]/1000)**3,color=colors[0])
plt.scatter(temps,(truesigma[:,1]/1000)**3,color=colors[1])
plt.scatter(temps,(truesigma[:,2]/1000)**3,color=colors[2])
plt.ylim(min((truesigma[:,0]/1000)**3),max((truesigma[:,0]/1000)**3))
plt.show()
```
