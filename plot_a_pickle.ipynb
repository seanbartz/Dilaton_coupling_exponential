```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ml=38
lambda1=-4.25
mu=128*8
df1=pd.read_pickle("data/sigma_transition_mq_"+str(ml)+"_lambda_"+str(lambda1)+"_mu_"+str(mu)+".pkl")

# Define the colors for each column, using the standard matplotlib color cycle
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
# for each element of the list sigma_list, plot the (sigma/1000)**3 vs. temperature and color the points according to the order
for i in range(len(df1['temps'])):
    for k in range(3):
        plt.scatter(df1['temps'][i],(df1['sigma'][i][:,k]/1000)**3,c=colors[k])
plt.xlabel("temperature")
plt.ylabel("$\sigma$ (GeV)$^3$")
#plt.ylim(min(df1['sigma'][0][:,0]/1000)**3,max(df1['sigma'][0][:,0]/1000)**3)
plt.title("mq="+str(ml)+" lambda="+str(lambda1)+" mu="+str(mu))
plt.show()

```
